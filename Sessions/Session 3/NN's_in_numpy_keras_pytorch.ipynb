### NN implementaion in numpy (one epoch)

```
import numpy as np
```

```
#Step0 : Inputs & Outputs
#X = raw_input("Enter the inputs of size (n,m)")
X = np.array(([1, 0, 1, 0], [1, 0, 1, 1], [0, 1, 0, 1]), dtype=float)
#Y = raw_input("Enter their respective outputs of size (1,m)")
Y = np.array(([1], [1], [0]), dtype=float).reshape(3,-1)
```

```
#Step1 : initializing weights and biases
wh = np.random.randn(4, 3) #hidden layer
wout = np.random.randn(3, 1) #output layer
bh = np.random.randn(1,3)
bout = np.random.randn(1, 1)
```

```
#Step2 : Calculate the hidden_layer_input
hidden_layer_input = X.dot(wh) + bh #linear function
```

```
#Step3 : Perform non-linear transformation(sigmoid) on hidden_linear_input to get hidden_layer_activations.
def sigmoid(x):
  return (1 / (1 + np.exp(x)))
hidden_layer_activations = sigmoid(hidden_layer_input)
```

```
#Step4 : Perform linear and non-linear transformation of hidden layer activation at output layer
output_layer_input = hidden_layer_activations.dot(wout) + bout
output = sigmoid(output_layer_input)
```

```
#Step5 : Calculate gradient of Error(E) at output layer
E = Y - output
```

```
#Step6: Compute slope at output and hidden layer (local gradients)
def derivatives_sigmoid(x):
  return (x * (1-x))
slope_output_layer = derivatives_sigmoid(output)
slope_hidden_layer = derivatives_sigmoid(hidden_layer_activations)
```

```
#Step7: Compute delta at output layer
lr = 0.1
d_output = E * slope_output_layer*lr
```

```
#Step8: Calculate Error at hidden layer
Error_at_hidden_layer = d_output * np.transpose(wout)
```

```
#Step9: Compute delta at hidden layer
d_hidden_layer = Error_at_hidden_layer * slope_hidden_layer
```

```
#Step10: Update weight at both output and hidden layer
wout = wout + (np.transpose(hidden_layer_activations).dot(d_output)) * lr
wh = wh+ (np.transpose(X).dot(d_hidden_layer)) * lr
```

```
#Step 11: Update biases at both output and hidden layer
bh = bh + np.sum(d_hidden_layer, axis=0) * lr
bout = bout + np.sum(d_output, axis=0)* lr
```

```
################
KERAS
################
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.optimizers import SGD
import numpy as np 
```

```
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
```

```
model_keras = Sequential()
model_keras.add(Dense(8, input_dim=2))
model_keras.add(Activation('relu'))
model_keras.add(Dense(1))
model_keras.add(Activation('sigmoid'))
```

```
model_keras.summary()
```

```
sgd = SGD(lr=0.1)
model_keras.compile(loss='binary_crossentropy', optimizer=sgd)
```

```
model_keras.fit(X, y, batch_size=1, nb_epoch=1000)
print(model_keras.predict_proba(X))
```

```
##################
PYTORCH
#################
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F	
import torch.optim as optim
from torch.autograd import Variable
```

```
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0,1,1,0]]).T
```

```
#have to pass tensors only
X_tensor = torch.FloatTensor(X)
y_tensor = torch.FloatTensor(y)
```

```
inputs = Variable(X_tensor)
labels = Variable(y_tensor, requires_grad=False)
```

```
model = nn.Sequential(nn.Linear(2, 8),    
                      nn.ReLU(),       
                      nn.Linear(8, 1),
                      nn.Sigmoid())
```

```
criterion = nn.BCELoss()
```

```
optimizer = optim.SGD(model.parameters(), lr=0.1)
num_epochs = 1000
```

```
for epoch in range(num_epochs):
    predictions = model(inputs)
    loss = criterion(predictions, labels)
    model.zero_grad()
    loss.backward()
    optimizer.step()
    if epoch%100 == 0: 
      print([float(_pred) for _pred in predictions], list(map(int, labels)), loss.data[0])
```
